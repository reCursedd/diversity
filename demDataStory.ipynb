# A Story of Hope

I looked around a bit and took up this very simple looking and cliched demographic data from [Data.gov](https://www.data.gov/), and it tells us about simple numbers surrounding the life expectancy at birth and death rates of US population, since 1900 till 2015. This is how the statistics of my data looks like: 

>(We first import the required pandas and matplotlib modules of python and called describe() to give stats of the entire data in hand.)


```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
 
file = r'mainData'
df = pd.read_csv(file)
print " Here 'Life' and 'Death' are aliases for Life Expectancy and Death Rates respectively"
df.describe()

```

## The Insights
I decided to look deeply into the ‘Life Expectancy’ column a bit more closely, expecting to see a smooth upward curve of a happy population. 

* **For the entire American population, this is what it looks like, rising as expected (in a roller coster ride), towards a better future.
**
> (loc() helps us in filtering the data very easily and we can generate the required plot.)

```python
all = df.loc[(df['Race'] == 'All Races')&(df['Sex'] == 'Both Sexes')] 
x = all['Year']
y = all['Life']
z = all['Death']
plt.plot(x,y,'green')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.savefig('life_all.png')
plt.show()
```

Remembering what really made the dent near 1916, I remembered the infamous World War I. But I also remembered that America was not as badly hit as Europe and decided to recheck my assumptions using the ‘Death Rate’ measure. And I was right!
* ** People died young and that somehow disturbed the Life Expectancy curve.**
* ** The death rates were not as alarming, just as expected.** 
> (Here x, y, z helps us the decide the axes needed)

```python
plt.plot(x,z,'red')
plt.xlabel('Year')
plt.ylabel('Death Rates')
plt.savefig('death_all.png')
plt.show()
```

Further, I went ahead and checked the distribution for different segments of the population, and here are some of the comparisons. The first one shows the life expectancy at birth of Female and Male population
* **No matter how women are oppressed, they outlive men and are naturally blessed with longer lives.**
* **Also, roughly the gap or the difference in life expectancy, of Females and Males is getting higher with time**

```python
fLife = df.loc[(df['Race'] == 'All Races')&(df['Sex'] == 'Female')] 
mLife = df.loc[(df['Race'] == 'All Races')&(df['Sex'] == 'Male')]
x = fLife['Year']
y = fLife['Life']
z = mLife['Life']
plt.plot(x,y,'magenta', label='Life Expectancy of Females')
plt.plot(x,z,'blue', label='Life Expectancy of Males')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()
plt.savefig('fm_all.png')
plt.show()
```

In the next plot, we plot the life expectancy and death rates of Black and White population.
We find that Blacks have lower life expectancy and higher death rates than White population. 
But in lieu of that, the hope lies in the fact that the curves are merging. 
* **Slowly but steadily the Blacks are facing lesser obstacles, to live. The death rates and life expectancy of Whites and Blacks are merging. 
**
>They are done in seperate plots for clarity, but subplot() function can be used to show two plots together. Here, fill_between() is used to simply fill in the region between 2 line plots. 

```python
ball = df.loc[(df['Race'] == 'Black')&(df['Sex'] == 'Both Sexes')] 
wall = df.loc[(df['Race'] == 'White')&(df['Sex'] == 'Both Sexes')]
x = ball['Year']
y = ball['Life']
z = wall['Life']
plt.plot(x,y,'maroon', label='Life Expectancy of Black population')
plt.plot(x,z,'pink', label='Life Expectancy of White population')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()
plt.fill_between(x,y,z, color = 'lightgreen')
plt.savefig('bw_life.png')
plt.show()
```

```python
ball = df.loc[(df['Race'] == 'Black')&(df['Sex'] == 'Both Sexes')] 
wall = df.loc[(df['Race'] == 'White')&(df['Sex'] == 'Both Sexes')]
x = ball['Year']
y = ball['Death']
z = wall['Death']
plt.plot(x,y,'maroon', label='Life Expectancy of Black population')
plt.plot(x,z,'pink', label='Life Expectancy of White population')
plt.xlabel('Year')
plt.ylabel('Death Rates')
plt.legend()
plt.fill_between(x,y,z, color = 'red')
plt.savefig('bw_death.png')
plt.show()
```
